# Seaborn 시각화

```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mpl.rcParams['axes.unicode_minus'] = False 
plt.rcParams["font.family"] = 'Malgun Gothic'
```

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

- countplot

```python
sns.countplot(x='class', data=titanic)
plt.title('타이타닉호의 각 클래스별 승객 수')
plt.show()
```

```python
sns.countplot(x="day", data=tips)
plt.title("요일별 팁을 준 횟수")
plt.show()
```

- pairplot

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.show()
```

- heatmap

```python
flights_passengers = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(8,6))
plt.title("연도, 월 별 승객수에 대한 Heatmap")
sns.heatmap(flights_passengers, annot=True, fmt="d", linewidths=1)
plt.show()
```

- barplot

```python
tips.head()
```

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("요일 별, 전체 요금의 평균")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", hue='sex', data=tips)
plt.title("요일별/성별 전체 요금의 평균")
plt.show()
```

- boxplot

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("요일 별, 전체 요금의 Box Plot")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue='sex', data=tips)
plt.title("요일별/성별 전체 요금의 Box Plot")
plt.show()
```
